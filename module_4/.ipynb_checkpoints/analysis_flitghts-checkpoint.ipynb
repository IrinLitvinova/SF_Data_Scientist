```python
# импортировать библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import re
%matplotlib inline

# загрузить файл
flights = pd.read_csv('D:/DS/UNIT4/flights_11122021_new1.csv') 
display(flights.sample(10))
display(flights.info())
'''
 Описание полей
 flight_id - номер рейса
 departure_airport - код аэропорта вылета
 arrival_airport - код аэропорта прибытия
 city - город прибытия
 actual_dearture - дата и время вылета
 day_departure - день недели вылета (0- вс)
 month_dep - месяц вылета
 actual_arrival - дата и время прибытия
 day_arrival - день недели прибытия (0- вс)
 model - модель самолета
 range - максимальная дальность полета
 time_fly - фактическое время в пути, минут
 tickets_count - количество проданных на рейс билетов
 seats_count - количество посадочных мест в можели самолета
 fullness - процент незаполненных мест на рейсе, %
 fullness = (количество посадочных мест - количество проданных билетов) /количество посадочных мест * 100%
 petrol_hour - нормативный расход топлива для модели самолета в час, л
 petrol_fly - фактический расход топлива за полет, л
 income - доход от продажи билетов на рейс, руб
 expense - расходы рейса на топливо, руб
 profit - валовая прибыль
'''
```

ОписаниеОпределяем полезные функции

```python
# определить размер датасета
flights.shape
# !!! Почему не 127 записей? - потому что я анализирую сезон - дек 2016 и янв,фев 2017
# В данный период схожие экономические условия, условия работы компании, также на практике обычно сравнивают
# данные одного сезона с другим, не только по календарному году
```

```python
flights.arrival_airport.unique() # куда летают самолеты из Анапы - Москва и Белгород
```

```python
# какие модели самолетов летают в оба аэропорта
flights.groupby(['arrival_airport']).model.unique()
```

```python
# рассчитаем процент заполняемости салона
flights['full_perc'] = round(flights.tickets_count / flights.seats_count * 100)
```

```python
# средний процент заполняемости по модели самолета
flights.groupby(['arrival_airport']).full_perc.mean()
```

```python
# заполняемость по дням недели
pivot_day = flights.pivot_table(values=['full_perc'],
index=['day_departure'],
columns=['arrival_airport'],
aggfunc='mean')
display(round(pivot_day,1))
```

```python
# заполняемость по месяцам
pivot_month = flights.pivot_table(values=['full_perc'],
index=['month_dep'],
columns=['arrival_airport'],
aggfunc='mean')
display(round(pivot_month),1)

```

```python
# количество проданных билетов на Боинг до Москвы (мест 130)
flights[flights.model == 'Boeing 737-300'].tickets_count.plot(kind = 'hist')
# рейс на Белгород практически всегда загружен полностью, на Москву наиболее частые продажи 105-110-117-130
# стоит более подробно посмотреть рейс на Москву, например, по дням неделям
# возможно, в какие-то дни запускать Суперджет на 97 мест, в какие-то Боинг на 130
```

```python
# количество проданных билетов на Суперджет до Беогорода(мест 97)
flights[flights.model == 'Sukhoi Superjet-100'].tickets_count.plot(kind = 'hist')
```

```python
# Суперджет практически всегда заполнен полностью, в отличие от Боинга
```

```python
# посмотрим какие рейсы наименее прибыльные
min_profit = flights.sort_values(by='profit', ascending=True).head(15)
display(min_profit)
```

```python
# сгруппируем их по дням недели
min_profit.groupby(['day_departure'])['day_departure'].count()
```

```python
# в вс и пн заполняемость небольшая, можно просчитать как пустить один большой самолет в один из дней вместо двух
```

```python
# рассчитаем отношение валовой прибыли к выручке
flights['profitability'] = round(flights.profit / flights.income *100,1)
display(flights.head(5))
```

```python
# В данном случае прибыльность имеет смысл рассматривать только сравнивая между собой рейсы двух моделей самолетов
# но не рейсы между собой, т.к. у одной модели самолетов расходы будут равны
```

```python
# рассчитаем среднюю прибыльность моделей
flights.groupby(['model'])['profitability'].mean()
```
